```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("asaniczka/top-spotify-songs-in-73-countries-daily-updated")

print("Path to dataset files:", path)
```

```python
import pandas as pd

df = pd.read_csv(path + "/universal_top_spotify_songs.csv")
df.head(10)
```

```python
df_1 = df.drop(['snapshot_date', 'album_name', 'album_release_date'], axis = 1)
```

```python
sample = df_1

sample['pop_trend_index'] = (1/sample['daily_rank']) + (sample['daily_movement']/10) + (sample['weekly_movement']/10) + (sample['popularity']/100)
sample = sample.drop(['daily_rank', 'daily_movement', 'weekly_movement', 'popularity'], axis = 1)
sample.info()
```

```python
kaggle_filtered = sample[['name', 'artists','energy', 'loudness', 'danceability', 'duration_ms', 'valence','instrumentalness', 'tempo', 'pop_trend_index']] ## resetpoint
kaggle_filtered.info()
kaggle_filtered.head(20)
```

```python
kaggle_filtered.info()
```

```python
kaggle_grouped = kaggle_filtered.groupby(['name', 'artists']).mean()
kaggle_popularity = kaggle_filtered[['name', 'artists', 'pop_trend_index']].groupby(['name', 'artists']).sum()
```

```python
kaggle_grouped = kaggle_grouped.reset_index()
kaggle_popularity = kaggle_popularity.reset_index()
kaggle_grouped.info()
```

```python
kaggle_grouped['pop_trend_index'] = kaggle_popularity['pop_trend_index']
kaggle_filtered = kaggle_grouped
kaggle_filtered.head(10)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler1 = StandardScaler()
scaler2 = MinMaxScaler()
kaggle_minmax = pd.DataFrame(scaler2.fit_transform(kaggle_filtered[['energy', 'danceability', 'valence','instrumentalness']]), columns = ['energy', 'danceability', 'valence','instrumentalness'])
kaggle_standard = pd.DataFrame(scaler1.fit_transform(kaggle_filtered[['loudness', 'duration_ms', 'tempo', 'pop_trend_index']]), columns = ['loudness', 'duration_ms', 'tempo', 'pop_trend_index'])
kaggle_filtered = pd.concat([sample[['name', 'artists']], kaggle_minmax, kaggle_standard], axis = 1)
```
